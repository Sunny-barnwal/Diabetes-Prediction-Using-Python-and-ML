```python
X = diabetes.drop('CLASS', axis=1)
X=X.drop('Gender',axis=1)
```

```python
Y = diabetes['CLASS']
```

```python
from sklearn.model_selection import train_test_split  #package for dividing the dataset into train and test
from sklearn.preprocessing import StandardScaler
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
scaler = StandardScaler()
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
scaler.fit(X_train)
```

```python
from sklearn.neighbors import KNeighborsClassifier #package for KNN algorithm
knn=KNeighborsClassifier(n_neighbors=9) #model building
knn.fit(X_train,Y_train) #fit method is used for training the model
```

```python
y_pred=knn.predict(X_test) #testing the model using predict method
print(y_pred)
```

```python
from sklearn.metrics import accuracy_score #package for finding accuracy
accuracy=accuracy_score(Y_test,y_pred) #accuracy_score(actual,predicted)
print(accuracy)
```
